# Approval rate by key categories

```python
import plotly.express as px

# Convert Loan_Status into numeric
# Y = Approved (1), N = Rejected (0)
df["Loan_Status_Num"] = df["Loan_Status"].map({"Y": 1, "N": 0})

# Function to calculate approval rate and plot
def approval_rate_plot(column_name):
    result = (
        df.groupby(column_name, dropna=False)["Loan_Status_Num"]
        .mean()
        .reset_index()
    )

    result["Approval_Rate_%"] = (result["Loan_Status_Num"] * 100).round(2)

    fig = px.bar(
        result,
        x=column_name,
        y="Approval_Rate_%",
        text="Approval_Rate_%",
        title=f"Approval Rate by {column_name}",
        labels={
            column_name: column_name,
            "Approval_Rate_%": "Approval Rate (%)"
        }
    )

    fig.update_traces(textposition="outside")
    fig.update_layout(
        yaxis_range=[0, 100],
        xaxis_title=column_name,
        yaxis_title="Approval Rate (%)"
    )

    fig.show()

    return result[[column_name, "Approval_Rate_%"]]

# Approval rates by important categories
approval_rate_plot("Gender")
approval_rate_plot("Married")
approval_rate_plot("Education")
approval_rate_plot("Self_Employed")
approval_rate_plot("Property_Area")
approval_rate_plot("Credit_History")
approval_rate_plot("Dependents")
```

# Credit history — dominant predictor

```python
import plotly.express as px

# Convert Loan_Status into numeric
df["Loan_Status_Num"] = df["Loan_Status"].map({"Y": 1, "N": 0})

# Credit history summary
credit_summary = (
    df.groupby("Credit_History")
    .agg(
        Applicants=("Loan_Status_Num", "count"),
        Approved=("Loan_Status_Num", "sum"),
        Approval_Rate=("Loan_Status_Num", "mean")
    )
    .reset_index()
)

# Convert to percentage
credit_summary["Approval_Rate_%"] = (
    credit_summary["Approval_Rate"] * 100
).round(2)

# Display summary table
display(credit_summary)

# ============================
# Plotly Bar Chart
# ============================

fig = px.bar(
    credit_summary,
    x="Credit_History",
    y="Approval_Rate_%",
    text="Approval_Rate_%",
    color="Credit_History",
    title="Loan Approval Rate by Credit History",
    labels={
        "Credit_History": "Credit History",
        "Approval_Rate_%": "Approval Rate (%)"
    }
)

fig.update_traces(textposition="outside")

fig.update_layout(
    yaxis_range=[0, 100],
    xaxis=dict(type="category")
)

fig.show()

# ============================
# Applicants vs Approved
# ============================

fig2 = px.bar(
    credit_summary,
    x="Credit_History",
    y=["Applicants", "Approved"],
    barmode="group",
    title="Applicants vs Approved by Credit History",
    labels={
        "value": "Count",
        "Credit_History": "Credit History"
    }
)

fig2.show()
```

# Income, loan amount & loan term patterns

```python
import plotly.express as px
import plotly.graph_objects as go

# Convert Loan_Status into numeric
df["Loan_Status_Num"] = df["Loan_Status"].map({"Y": 1, "N": 0})

# Create Total Income column
df["TotalIncome"] = (
    df["ApplicantIncome"] + df["CoapplicantIncome"]
)

# Income and Loan Summary
income_loan_summary = df.groupby("Loan_Status").agg(
    Applicants=("Loan_Status", "size"),
    Avg_Applicant_Income=("ApplicantIncome", "mean"),
    Median_Applicant_Income=("ApplicantIncome", "median"),
    Avg_Coapplicant_Income=("CoapplicantIncome", "mean"),
    Avg_Total_Income=("TotalIncome", "mean"),
    Median_Total_Income=("TotalIncome", "median"),
    Avg_Loan_Amount=("LoanAmount", "mean"),
    Median_Loan_Amount=("LoanAmount", "median"),
    Avg_Loan_Term=("Loan_Amount_Term", "mean"),
    Median_Loan_Term=("Loan_Amount_Term", "median")
).round(2)

# Display summary table
display(income_loan_summary)

# Reset index for plotting
summary_df = income_loan_summary.reset_index()

# =====================================
# Average Applicant Income
# =====================================

fig1 = px.bar(
    summary_df,
    x="Loan_Status",
    y="Avg_Applicant_Income",
    text="Avg_Applicant_Income",
    color="Loan_Status",
    title="Average Applicant Income by Loan Status"
)

fig1.update_traces(textposition="outside")
fig1.show()

# =====================================
# Average Total Income
# =====================================

fig2 = px.bar(
    summary_df,
    x="Loan_Status",
    y="Avg_Total_Income",
    text="Avg_Total_Income",
    color="Loan_Status",
    title="Average Total Income by Loan Status"
)

fig2.update_traces(textposition="outside")
fig2.show()

# =====================================
# Average Loan Amount
# =====================================

fig3 = px.bar(
    summary_df,
    x="Loan_Status",
    y="Avg_Loan_Amount",
    text="Avg_Loan_Amount",
    color="Loan_Status",
    title="Average Loan Amount by Loan Status"
)

fig3.update_traces(textposition="outside")
fig3.show()

# =====================================
# Loan Term Comparison
# =====================================

fig4 = px.bar(
    summary_df,
    x="Loan_Status",
    y="Avg_Loan_Term",
    text="Avg_Loan_Term",
    color="Loan_Status",
    title="Average Loan Term by Loan Status"
)

fig4.update_traces(textposition="outside")
fig4.show()

# =====================================
# Applicants Count
# =====================================

fig5 = px.pie(
    summary_df,
    names="Loan_Status",
    values="Applicants",
    title="Applicants Distribution by Loan Status",
    hole=0.4
)

fig5.show()
```

# Feature correlations with approval

```python
import plotly.express as px
import pandas as pd

# Convert target to numeric
df["Loan_Status_Num"] = df["Loan_Status"].map({"Y": 1, "N": 0})

# Create Total Income feature
df["TotalIncome"] = (
    df["ApplicantIncome"] + df["CoapplicantIncome"]
)

# Select numeric columns
numeric_df = df.select_dtypes(include=["int64", "float64"])

# Correlation with Loan Approval
approval_corr = (
    numeric_df.corr()["Loan_Status_Num"]
    .sort_values(ascending=False)
)

# Convert to dataframe
approval_corr_df = approval_corr.reset_index()
approval_corr_df.columns = ["Feature", "Correlation"]

# Display correlation table
display(approval_corr_df)

# =========================================
# Plotly Correlation Bar Chart
# =========================================

fig = px.bar(
    approval_corr_df,
    x="Feature",
    y="Correlation",
    text="Correlation",
    color="Correlation",
    title="Feature Correlation with Loan Approval"
)

fig.update_traces(
    texttemplate='%{text:.2f}',
    textposition='outside'
)

fig.update_layout(
    xaxis_title="Features",
    yaxis_title="Correlation Value",
    xaxis_tickangle=-45
)

fig.show()

# =========================================
# Heatmap of Numeric Correlations
# =========================================

corr_matrix = numeric_df.corr().round(2)

fig2 = px.imshow(
    corr_matrix,
    text_auto=True,
    aspect="auto",
    title="Correlation Heatmap of Numeric Features"
)

fig2.show()
```
